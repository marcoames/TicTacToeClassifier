```python
import random
import pandas as pd

# Função para verificar se há um vencedor em um tabuleiro
def check_winner(board, player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Linhas
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Colunas
        [0, 4, 8], [2, 4, 6]  # Diagonais
    ]
    return any(all(board[i] == player for i in condition) for condition in win_conditions)

# Função para classificar o tabuleiro
def classify_board(board):
    if check_winner(board, 'x'):
        return 'Xganha'
    if check_winner(board, 'o'):
        return 'Oganha'
    if 'b' in board:
        return 'Temjogo'
    return 'Empate'

# Verifica se o tabuleiro é válido (número de 'x' e 'o')
def is_valid_board(board):
    count_x = board.count('x')
    count_o = board.count('o')
    # X deve ter um a mais ou igual que O
    return count_x >= count_o and count_x <= count_o + 1

# Gerar um número limitado de tabuleiros
def generate_sample_boards(num_samples):
    possible_values = ['x', 'o', 'b']
    boards = []
    results = []

    while len(boards) < num_samples:
        board = [random.choice(possible_values) for _ in range(9)]
        if is_valid_board(board):  # Verifica se o tabuleiro é válido
            board_result = classify_board(board)
            boards.append(board)
            results.append(board_result)
    
    return boards, results


# Gerar tabuleiros e resultados
boards, results = generate_sample_boards(10000)

# Criar DataFrame
data = pd.DataFrame(boards, columns=['1', '2', '3', '4', '5', '6', '7', '8', '9'])
data['result'] = results

# Remover duplicatas
data = data.drop_duplicates()

# Contar as instâncias de cada resultado
print(data['result'].value_counts())

# Salvar o DataFrame em um arquivo CSV
data.to_csv('data/dataset.csv', index=False)

# Visualizar as primeiras linhas do DataFrame
print(len(data))
print(data.head(10))
```
